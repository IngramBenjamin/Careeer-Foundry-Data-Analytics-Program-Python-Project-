```python
# In this Exercise, you learned how to find the aggregated mean of the “order_number” column grouped by “department_id” for a subset of your dataframe. Now, repeat this process for the entire dataframe.
ords_prods_merge.groupby('department_id').agg({'number_of_orders_placed': ['mean']})
```

```python
# Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

# Creating a max order column grouped by user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['number_of_orders_placed'].transform(np.max)
```

```python
# Changing pandas display setting to not cap rows displayed 
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(100)
```

```python
# Using loc statement to create loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'number_of_orders_placed']].head(60)
```

```python
ords_prods_merge['loyalty_flag'].value_counts()
```

```python
# The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). 

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders

# Creating an average price column grouped by user
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(100)
```

```python
# Using loc statement to create spender flag
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spender_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
# Checking frequency of spender flag
ords_prods_merge['spender_flag'].value_counts()
```

```python
# Checking accuracy of high spender flag assignment 
ords_prods_merge.query('avg_price >= 10').head()
```

```python
# Checking accuracy of low spender flag assignment 
ords_prods_merge.query('avg_price < 10').head()
```

```python
#In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

# Creating an median order frequency column grouped by user
ords_prods_merge['med_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(100)
```

```python
# Using loc statement to create frequent shopper flag
ords_prods_merge.loc[ords_prods_merge['med_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['med_frequency'] <= 20) & (ords_prods_merge['med_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['med_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking frequency of frequent shopper flag
ords_prods_merge['frequency_flag'].value_counts()
```

```python
# Checking accuracy of frequency flag assignment 
ords_prods_merge.head(100)
```

```python
# Replacing price outliers with missing values 

ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
