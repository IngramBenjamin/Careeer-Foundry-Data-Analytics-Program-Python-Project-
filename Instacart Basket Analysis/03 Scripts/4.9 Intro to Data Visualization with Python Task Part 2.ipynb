# 01. Import Libraries and Data

```python
# Importing libraries 
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Importing orders, products, and customers merged dataframes 

path = r"C:\Users\maple\OneDrive\Desktop\CareerFoundry\Instacart Basket Analysis"
ords_prods_custs = pd.read_pickle(os.path.join(path,'02 Data','Prepared Data', 'orders_products_customers_merged.pkl'))
```

# 02. Data Visualization Exercises

```python
# Creating bar chart of orders_days_of_week 

bar = ords_prods_custs['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Creating histogram of prices

histogram = ords_prods_custs['prices'].plot.hist(bins = 70)
```

```python
# Creating scatterplot of prices

scatterplot = sns.scatterplot(x = 'prices', y = 'prices', data = ords_prods_custs)
```

```python
# Create a list holding True/False values to the test np.random.rant() <= .7

np.random.seed(4)
dev = np.random.rand(len(ords_prods_custs)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big

big = ords_prods_custs[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = ords_prods_custs[~dev]
```

```python
# Creating subset of data needed for line chart

df_2 = small[['orders_day_of_week','prices']]
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

# 03. Histogram of order_hour_of_day

```python
# Creating histogram of order_hour_of_day

histogram2 = ords_prods_custs['order_hour_of_day'].plot.hist(bins = 24)
```

#### The histogram demonstrates that very few orders are placed during early morning hours and begin to increase around 7am.  Most orders take place between 10am and 4pm, and orders begin to decrease in the evening into the night. 

# 04. Bar chart from loyalty_flag

```python
# Creating bar chart of orders_days_of_week 

bar2 = ords_prods_custs['loyalty_flag'].value_counts().plot.bar()
```

# 05. Line chart prices and hour of day

```python
# Creating subset of data needed for line chart

df_3 = small[['order_hour_of_day','prices']]
```

```python
line2 = sns.lineplot(data = df_3, x = 'order_hour_of_day',y = 'prices')
```

#### Prices do not vary greatly across hours of the day but do drop moderately around 1-2am and again at 10am 

# 06. Line chart age and number of dependents

```python
# Creating subset of data needed for line chart
df_4 = small[['age','number_of_dependents']]
```

```python
line3 = sns.lineplot(data = df_4, x = 'age',y = 'number_of_dependents')
```

#### The line chart shows that there is no strong relationship between age and number of dependents. The number of dependents flucutates frequently with values between 1.44 and 1.58 across age ranges, with there being no indication that the number of dependents is not higher or lower for older age groups when compared to younger age groups

# 07. Scatterplot of age and income

```python
# Creating scatterplot of age and income

scatterplot2 = sns.scatterplot(x = 'age', y = 'income', data = ords_prods_custs)
```

#### The incomes of customers aged 40 and younger are lower than customers older than 40. This is likely because older customers are more established in their careers having spent more time in the workforce
